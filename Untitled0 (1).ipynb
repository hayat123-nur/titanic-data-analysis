```python


file_path = "/content/drive/MyDrive/dataset-2.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
print("\n INFO ")
print(df.info())

print("\n DESCRIPTION ")
print(df.describe())
```

```python
print("\n MISSING VALUES BEFORE ")
print(df.isnull().sum())

# Fill missing values (clean way)
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop Cabin safely
if "Cabin" in df.columns:
    df.drop("Cabin", axis=1, inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

print("\n MISSING VALUES AFTER ")
print(df.isnull().sum())
```

```python
print("\n SURVIVAL RATE BY GENDER ")
print(df.groupby("Sex")["Survived"].mean())
```

```python
print("\n SURVIVAL RATE BY CLASS ")
print(df.groupby("Pclass")["Survived"].mean())
```

```python
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 60, 100],
    labels=["Child", "Teen", "Adult", "Old"]
)

print("\n SURVIVAL RATE BY AGE GROUP ")
print(df.groupby("AgeGroup")["Survived"].mean())
```

```python
print("\nFEMALES WHO SURVIVED ")
print(df[(df["Sex"] == "female") & (df["Survived"] == 1)])

print("\n CHILDREN WHO SURVIVED ")
print(df[(df["Age"] < 12) & (df["Survived"] == 1)])

print("\n 1ST CLASS PASSENGERS WHO SURVIVED ")
print(df[(df["Pclass"] == 1) & (df["Survived"] == 1)])
```

```python
print("\n INSIGHTS ")

print("1. Females were more likely to survive than males.")
print("2. Passenger class affected survival; 1st class had the highest survival rate.")
print("3. Children had higher survival chances than adults.")
print("4. Female passengers in 1st class had the highest survival rate.")
```
